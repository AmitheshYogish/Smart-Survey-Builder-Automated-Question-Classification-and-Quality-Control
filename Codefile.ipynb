   - Loads tokenizer (`question_tokenizer`) and label encoder (`label_encoder`) used with the LSTM model.
   - Initializes a grammar correction model (`grammar_corrector`) using T5-based model from Hugging Face's Transformers library.
   - Initializes an emotion classification pipeline (`tone_classifier`) using a DistilRoBERTa-based model for classifying emotions in text.
   - Initializes a profanity detector using `better_profanity` library.

# Models Used:

### 1. Attention-based LSTM Model (`question_model`):

- **Purpose**: This model is used for question classification, which involves categorizing a given question into predefined classes or types (e.g., rating scale questions, open-ended questions).
  
- **Training**:
  - **Architecture**: Long Short-Term Memory (LSTM) networks are a type of recurrent neural network (RNN) designed to capture long-term dependencies in sequential data.
  - **Training Data**: Typically trained on a labeled dataset where questions are annotated with their respective types.
  - **Loss Function**: Categorical cross-entropy loss is commonly used for multi-class classification tasks like question type classification.
  - **Optimization**: Optimized using the Adam optimizer, which adapts learning rates for each parameter.
  - **Metrics**: Evaluated using accuracy and confusion matrix to evaluate its performance

### 2. T5-Based Grammar Correction Model (`grammar_corrector`):

- **Purpose**: This model corrects grammar errors in text input, improving readability and correctness.
  
- **Training**:
  - **Architecture**: Uses T5 (Text-To-Text Transfer Transformer), a transformer-based model capable of performing various text-to-text tasks, including grammar correction.
  - **Training Data**: Trained on a large corpus of text with labeled examples of incorrect and corrected sentences.
  - **Loss Function**: Trained using masked language modeling and sequence-to-sequence objectives.
  - **Fine-tuning**: Fine-tuned on specific grammar correction tasks to adapt to domain-specific or language-specific nuances.

### 3. DistilRoBERTa-Based Emotion Classification Model (`tone_classifier`):

- **Purpose**: This model predicts the emotional tone of a given text, identifying emotions such as joy, sadness, anger, etc.
  
- **Training**:
  - **Architecture**: Based on DistilRoBERTa, a distilled version of RoBERTa (a variant of BERT), optimized for efficiency while maintaining high performance.
  - **Training Data**: Trained on a dataset with text samples labeled with emotional categories.
  - **Loss Function**: Typically uses categorical cross-entropy loss for multi-class emotion classification.
  - **Fine-tuning**: Fine-tuned on emotion classification tasks to capture nuanced emotional cues in text.

### 4. Profanity Detection

- **Purpose**: This pipeline detects and masks profanity in textual data, ensuring content meets standards of appropriateness.
  
- **Preprocessing**:
  - **Lowercasing**: Converts all text to lowercase for uniform processing.
  - **Special Character Removal**: Eliminates non-alphanumeric characters while preserving spaces.
  - **Number Conversion**: Replaces numeric digits with textual representations.
  - **Whitespace Normalization**: Reduces multiple spaces to single spaces.

- **Profanity Detection**:
  - **Initialization**: Loads profanity word list and necessary libraries.
  - **Text Preprocessing**: Applies preprocessing steps to prepare text for profanity detection.
  - **Profanity Check**: Utilizes a profanity detection library to identify if profanity exists in the preprocessed text.
  - **Censorship**: Generates a censored version of the text, replacing profane words with asterisks (`*`).
  - **Tokenization**: Splits the original and censored texts into tokens for comparison.
  - **Profane Word Identification**: Compares original and censored tokens to pinpoint profane words.

### 5. BERT-Based Spam Detection Model (`spam_classifier`):

- **Purpose**: This model identifies spam and phishing messages in text, crucial for filtering out unwanted or malicious content.
  
- **Training**:
  - **Architecture**: Based on BERT (Bidirectional Encoder Representations from Transformers), a transformer-based model known for its effectiveness in natural language understanding tasks.
  - **Training Data**: Trained on a dataset containing examples of SMS or text messages labeled as spam or not spam.
  - **Loss Function**: Binary cross-entropy loss is typically used for binary classification tasks like spam detection.
  - **Fine-tuning**: Fine-tuned on the specific task of spam detection to improve accuracy and adapt to various types of spam messages.


```python
import pandas as pd
import numpy as np
import re
import pickle
from better_profanity import profanity
import nltk
from nltk.tokenize import word_tokenize
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM, TFAutoModelForSequenceClassification
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Download necessary NLTK data
nltk.download('punkt', quiet=True)

# Load the attention-based LSTM model, tokenizer, and label encoder
question_model = load_model('/content/drive/MyDrive/question_classification_model_lstm.h5', compile=False)
question_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

with open('/content/drive/MyDrive/question_tokenizer_lstm.pickle', 'rb') as handle:
    question_tokenizer = pickle.load(handle)
with open('/content/drive/MyDrive/label_encoder_lstm.pickle', 'rb') as handle:
    label_encoder = pickle.load(handle)

# Initialize the grammar correction model
grammar_tokenizer = AutoTokenizer.from_pretrained("vennify/t5-base-grammar-correction")
grammar_model = AutoModelForSeq2SeqLM.from_pretrained("vennify/t5-base-grammar-correction")
grammar_corrector = pipeline("text2text-generation", model=grammar_model, tokenizer=grammar_tokenizer)

# Initialize the emotion classification pipeline
tone_classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", top_k=3)

# Initialize the profanity detector
profanity.load_censor_words()

# Initialize the spam detection model
spam_tokenizer = AutoTokenizer.from_pretrained("mrm8488/bert-tiny-finetuned-sms-spam-detection")
spam_model = TFAutoModelForSequenceClassification.from_pretrained("mrm8488/bert-tiny-finetuned-sms-spam-detection")
spam_classifier = pipeline("text-classification", model=spam_model, tokenizer=spam_tokenizer)

# Constants
MAX_LEN = 100  # Make sure this matches the value used during training

# Helper functions
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = ' '.join(text.split())
    return text

def classify_question_type(question):
    sequence = question_tokenizer.texts_to_sequences([question])
    padded_sequence = pad_sequences(sequence, maxlen=MAX_LEN)
    prediction = question_model.predict(padded_sequence)
    predicted_label = label_encoder.inverse_transform([np.argmax(prediction)])
    return predicted_label[0]

def correct_grammar(text):
    corrected_question = grammar_corrector(text, max_length=128)[0]['generated_text']
    corrected_question = corrected_question.capitalize()

    proper_nouns = ["Facebook", "Twitter", "Instagram", "LinkedIn", "United States", "Chinese", "Mexican", "Indian", "Italian"]
    for noun in proper_nouns:
        corrected_question = re.sub(rf'\b{noun.lower()}\b', noun, corrected_question, flags=re.IGNORECASE)

    return corrected_question

def analyze_tone(text):
    preprocessed_text = preprocess_text(text)
    results = tone_classifier(preprocessed_text)
    return ", ".join([f"{result['label']} ({result['score']:.2f})" for result in results[0]])

def detect_profanity(text):
    contains_profanity = profanity.contains_profanity(text)
    censored_text = profanity.censor(text)
    original_words = text.split()
    censored_words = censored_text.split()
    profane_words = [orig for orig, cens in zip(original_words, censored_words) if '*' in cens]
    return {
        'contains_profanity': contains_profanity,
        'censored_text': censored_text,
        'profane_words': profane_words
    }

# Update the detect_spam function
def detect_spam(text):
    result = spam_classifier(text)[0]
    label = "spam/phishing" if result['label'] == 'LABEL_1' or result['score'] > 0.5 else "not spam"
    return label, result['score']

# Main pipeline function
def process_question(question_data):
    if isinstance(question_data, tuple):
        question = question_data[0]
        options = question_data[1:]
    else:
        question = question_data
        options = []

    # Step 1: Question Type Classification
    question_type = classify_question_type(question)

    # Step 2: Grammar Correction
    corrected_question = correct_grammar(question)

    # Step 3: Tone Analysis (using corrected question)
    tone = analyze_tone(corrected_question)

    # Step 4: Profanity Detection and Masking (using corrected question)
    profanity_result = detect_profanity(corrected_question)
    clean_question = profanity_result['censored_text']

    # Step 5: Spam Detection (using corrected question)
    spam_label, spam_score = detect_spam(corrected_question)

    # Interpret spam results
    if spam_label == "spam/phishing" or spam_score > 0.5:
        spam_status = f"spam/phishing (score: {spam_score:.2f})"
    else:
        spam_status = f"not spam (score: {spam_score:.2f})"

    # Compile Results
    result = {
        "original_question": question,
        "options": options,
        "question_type": question_type,
        "corrected_question": corrected_question,
        "clean_question": clean_question,
        "tone": tone,
        "contains_profanity": profanity_result['contains_profanity'],
        "profane_words": profanity_result['profane_words'],
        "spam_status": spam_status
    }

    return result
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Attention, GlobalAveragePooling1D
from tensorflow.keras.callbacks import EarlyStopping
import nltk
from nltk.corpus import wordnet
import random
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Download WordNet for synonym replacement
nltk.download('wordnet')

# Load your dataset
data = pd.read_csv('/content/drive/MyDrive/survey_questions_spam_rand.csv')  # Replace with your actual file path
X = data['Question'].values
y = data['Type'].values

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
num_classes = len(le.classes_)

# Function to perform synonym replacement
def synonym_replacement(text, n=1):
    words = text.split()
    new_words = words.copy()
    random_word_list = list(set([word for word in words if word.isalnum()]))
    random.shuffle(random_word_list)
    num_replaced = 0
    for random_word in random_word_list:
        synonyms = []
        for syn in wordnet.synsets(random_word):
            for l in syn.lemmas():
                synonyms.append(l.name())
        if len(synonyms) >= 1:
            synonym = random.choice(list(set(synonyms)))
            new_words = [synonym if word == random_word else word for word in new_words]
            num_replaced += 1
        if num_replaced >= n:
            break
    return ' '.join(new_words)

# Function to augment data
def augment_data(X, y, augment_factor=2):
    augmented_X = []
    augmented_y = []
    for text, label in zip(X, y):
        augmented_X.append(text)
        augmented_y.append(label)
        for _ in range(augment_factor - 1):
            augmented_text = synonym_replacement(text)
            augmented_X.append(augmented_text)
            augmented_y.append(label)
    return np.array(augmented_X), np.array(augmented_y)

# Augment the data
X_augmented, y_augmented = augment_data(X, y_encoded)

# Tokenize the text
max_words = 5000
max_len = 100

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_augmented)

X_sequences = tokenizer.texts_to_sequences(X_augmented)
X_padded = pad_sequences(X_sequences, maxlen=max_len)

# Convert labels to categorical
y_categorical = to_categorical(y_augmented)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_categorical, test_size=0.2, random_state=42)

# Function to create attention LSTM model
def create_attention_lstm_model(max_words, max_len, num_classes):
    inputs = Input(shape=(max_len,))
    embedding = Embedding(input_dim=max_words, output_dim=100, input_length=max_len)(inputs)
    lstm = LSTM(128, return_sequences=True)(embedding)
    attention = Attention()([lstm, lstm])
    avg_pool = GlobalAveragePooling1D()(attention)
    outputs = Dense(num_classes, activation='softmax')(avg_pool)
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Create and train the model
model = create_attention_lstm_model(max_words, max_len, num_classes)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(X_train, y_train, epochs=18, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# Evaluate the model
train_loss, train_accuracy = model.evaluate(X_train, y_train, verbose=0)
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)

print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Predictions
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)
```

  - **Training Data**: Trained on a large corpus of text with labeled examples of incorrect and corrected sentences.
  - **Loss Function**: Trained using masked language modeling and sequence-to-sequence objectives.

```python
import pandas as pd
import re
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/survey_questions_spam_rand.csv')

# Initialize the grammar correction model
tokenizer = AutoTokenizer.from_pretrained("vennify/t5-base-grammar-correction")
model = AutoModelForSeq2SeqLM.from_pretrained("vennify/t5-base-grammar-correction")
grammar_corrector = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

def preprocess_and_correct(text):
    # Split the question and options
    parts = re.split(r'(\([a-z]\))', text)
    question = parts[0].strip()
    options = parts[1:] if len(parts) > 1 else []

    # Correct only the question part
    corrected_question = grammar_corrector(question, max_length=128)[0]['generated_text']

    # Capitalize the first letter and proper nouns
    corrected_question = corrected_question.capitalize()
    proper_nouns = ["Facebook", "Twitter", "Instagram", "LinkedIn", "United States", "Chinese", "Mexican", "Indian", "Italian"]
    for noun in proper_nouns:
        corrected_question = re.sub(rf'\b{noun.lower()}\b', noun, corrected_question, flags=re.IGNORECASE)

    # Reconstruct the full question with options
    full_corrected = corrected_question
    for i in range(0, len(options), 2):
        if i+1 < len(options):
            full_corrected += f" {options[i]}{options[i+1]}"

    return full_corrected

# Apply grammar correction to each question
df['Corrected_Question'] = df['Question'].apply(preprocess_and_correct)

# Print some examples
print(f"Total questions: {len(df)}")
for _, row in df.head(10).iterrows():
    print("\nOriginal Question:", row['Question'])
    print("Corrected Question:", row['Corrected_Question'])

# Save the new dataset with corrected questions
#df.to_csv('D:\AMITHESH\survey_questions_grammar_corrected_tf1.csv', index=False)
#print("\nNew dataset with corrected questions has been saved as 'survey_questions_grammar_corrected.csv'")
```

## CODE SNIPPET FOR TONE ANALYSIS

### DistilRoBERTa-Based Emotion Classification Model (`tone_classifier`):

- **Purpose**: This model predicts the emotional tone of a given text, identifying emotions such as joy, sadness, anger, etc.
  
- **Training**:
  - **Architecture**: Based on DistilRoBERTa, a distilled version of RoBERTa (a variant of BERT), optimized for efficiency while maintaining high performance.
  - **Training Data**: Trained on a dataset with text samples labeled with emotional categories.
  - **Loss Function**: Typically uses categorical cross-entropy loss for multi-class emotion classification.
  - **Fine-tuning**: Fine-tuned on emotion classification tasks to capture nuanced emotional cues in text.

```python
import pandas as pd
import re
from transformers import pipeline
from tqdm import tqdm

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/survey_questions_spam_rand.csv')

# Initialize the emotion classification pipeline
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", top_k=3)

# Light preprocessing function
def preprocess_text(text):
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove URLs (if any)
    text = re.sub(r'http\S+|www.\S+', '', text)

    # Remove email addresses (if any)
    text = re.sub(r'\S+@\S+', '', text)

    # Replace multiple punctuation with single
    text = re.sub(r'([!?.]){2,}', r'\1', text)

    return text

# Function to analyze tone using BERT
def analyze_tone_bert(text):
    preprocessed_text = preprocess_text(text)
    results = classifier(preprocessed_text)
    return results[0]

# Apply tone analysis to each question
tqdm.pandas(desc="Analyzing tone")
df['Tone_Analysis'] = df['Question'].progress_apply(analyze_tone_bert)

# Function to format tone results
def format_tone_results(results):
    return ", ".join([f"{result['label']} ({result['score']:.2f})" for result in results])

# Format the tone results for better readability
df['Formatted_Tone'] = df['Tone_Analysis'].apply(format_tone_results)

# Print some examples
print(f"Total questions analyzed: {len(df)}")
for _, row in df.head(10).iterrows():
    print("\nOriginal Question:", row['Question'])
    print("Preprocessed Question:", preprocess_text(row['Question']))
    print("Tone Analysis:", row['Formatted_Tone'])

# Save the results
#df.to_csv('D:\AMITHESH\survey_questions_tone_analysis.csv', index=False)
#print("\nTone analysis results have been saved as 'survey_questions_tone_analysis.csv'")
```

## CODE SNIPPET FOR PROFANITY DETECTION:

### Profanity Detection Pipeline

- **Purpose**: This pipeline detects and masks profanity in textual data, ensuring content meets standards of appropriateness.
  
- **Preprocessing**:
  - **Lowercasing**: Converts all text to lowercase for uniform processing.
  - **Special Character Removal**: Eliminates non-alphanumeric characters while preserving spaces.
  - **Number Conversion**: Replaces numeric digits with textual representations.
  - **Whitespace Normalization**: Reduces multiple spaces to single spaces.

- **Profanity Detection**:
  - **Initialization**: Loads profanity word list and necessary libraries.
  - **Text Preprocessing**: Applies preprocessing steps to prepare text for profanity detection.
  - **Profanity Check**: Utilizes a profanity detection library to identify if profanity exists in the preprocessed text.
  - **Censorship**: Generates a censored version of the text, replacing profane words with asterisks (`*`).
  - **Tokenization**: Splits the original and censored texts into tokens for comparison.
  - **Profane Word Identification**: Compares original and censored tokens to pinpoint profane words.



```python
import pandas as pd
import re
from better_profanity import profanity
import nltk
from nltk.tokenize import word_tokenize

# Download necessary NLTK data
nltk.download('punkt', quiet=True)

# Initialize the profanity detector
profanity.load_censor_words()

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters, keeping spaces
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Replace numbers with their textual representation (simple version)
    text = text.replace('0', 'zero').replace('1', 'one').replace('3', 'three')
    text = text.replace('4', 'four').replace('5', 'five').replace('7', 'seven')

    # Remove extra spaces
    text = ' '.join(text.split())

    return text

def detect_profanity(text):
    # Preprocess the text
    preprocessed_text = preprocess_text(text)

    # Check for profanity in the preprocessed text
    contains_profanity = profanity.contains_profanity(preprocessed_text)

    # Get censored version of the text
    censored_text = profanity.censor(text)

    # Tokenize the original and censored text
    original_tokens = word_tokenize(text.lower())
    censored_tokens = word_tokenize(censored_text.lower())

    # Find the profane words
    profane_words = [orig for orig, cens in zip(original_tokens, censored_tokens) if orig != cens]

    return {
        'contains_profanity': contains_profanity,
        'censored_text': censored_text,
        'profane_words': profane_words
    }

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/survey_questions_spam_rand.csv')

# Apply profanity detection to each question
df['profanity_analysis'] = df['Question'].apply(detect_profanity)

# Extract profanity flag for easier viewing
df['Contains_Profanity'] = df['profanity_analysis'].apply(lambda x: x['contains_profanity'])

# Print summary of profanity analysis
print(f"Total questions: {len(df)}")
print(f"Questions containing profanity: {sum(df['Contains_Profanity'])}")

# Print examples of questions with profanity
print("\nExamples of questions containing profanity:")
for _, row in df[df['Contains_Profanity']].iterrows():
    print(f"\nOriginal Question: {row['Question']}")
    print(f"Question Type: {row['Type']}")
    print(f"Censored Question: {row['profanity_analysis']['censored_text']}")
    print(f"Profane words detected: {', '.join(row['profanity_analysis']['profane_words'])}")
    print("-" * 50)

```
